```python
import folium
import pandas as pd

eco_footprints = pd.read_csv("files/regions_stat.csv")
political_countries_url = ('files/russia_regions_new.geojson')

m = folium.Map(location=(30, 10), zoom_start=3, tiles="cartodb positron")
folium.Choropleth(
    geo_data=political_countries_url,
    data=eco_footprints,
    # Первый элемент ключ к-ый свяжет данные с данными GeoJSON
    columns=["Регион", "Количество"],
    key_on="feature.properties.name",
).add_to(m)

m.save("footprint.html")
```
